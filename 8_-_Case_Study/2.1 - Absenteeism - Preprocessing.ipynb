### .get_dumies()

```python
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns.head()
```

```python
reason_columns = reason_columns.drop(['check'], axis = 1)
reason_columns.head()
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
reason_columns.head()
```

### Group the Reasons for Absence:

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
df.head()
```

```python
reason_columns.loc[:, :14].head()
```

```python
reason_columns.loc[:, :14].max(axis = 1).head()
```

```python
Reason_type_1 = reason_columns.loc[:, :14].max(axis = 1)
Reason_type_2 = reason_columns.loc[:, 15:17].max(axis = 1)
Reason_type_3 = reason_columns.loc[:, 18:21].max(axis = 1)
Reason_type_4 = reason_columns.loc[:, 22:].max(axis = 1)
```

### Concatenate Column Values

```python
df.head()
```

```python
df = pd.concat([df, Reason_type_1, Reason_type_2, Reason_type_3, Reason_type_4], axis = 1)
df.head()
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
```

```python
df.head()
```

### Reorder Columns

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
df.head()
```

#### Create a Checkpoint

```python
df_reason_mod = df.copy()
df_reason_mod.head()
```

### 'Date':

```python
type(df_reason_mod['Date'][0])
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date']
```

```python
type(df_reason_mod['Date'][0])
```

#### Extract the Month Value:

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []
list_months
```

```python
df_reason_mod.shape
```

```python
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months[0:25]
```

```python
len(list_months)
```

```python
df_reason_mod['Mounth Value'] = list_months
df_reason_mod.head()
```

#### Extract the Day of the Week:

```python
df_reason_mod['Date'][0].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

#### Drop 'Date':

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
df_reason_mod.head()
```

```python
df_reason_mod.columns.values
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Mounth Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[column_names_reordered]
df_reason_mod.head()
```

#### Checkpoint!
