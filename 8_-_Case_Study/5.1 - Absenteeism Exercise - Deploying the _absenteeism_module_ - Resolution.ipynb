```python
predicted_df = model.predicted_outputs()
predicted_df.head()
```

## Analysis

```python
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
```

```python
x = predicted_df['Age']
y = predicted_df['Probability']*100

plt.scatter(x, y)

plt.title('Age vs. Probability')
plt.xlabel('Age')
plt.ylabel('Probability')

plt.show()
```

```python
x1 = predicted_df['Reason_1']
x2 = predicted_df['Reason_2']
x3 = predicted_df['Reason_3']
x4 = predicted_df['Reason_4']
y = predicted_df['Probability']*100

fig, axs = plt.subplots(1, 4, figsize=(9, 4), sharey=True)
axs[0].scatter(x1, y)
axs[0].set_xlabel('Reason_1')
axs[0].set_ylabel('Probability')
axs[1].scatter(x2, y)
axs[1].set_xlabel('Reason_2')
axs[2].scatter(x3, y)
axs[2].set_xlabel('Reason_3')
axs[3].scatter(x4, y)
axs[3].set_xlabel('Reason_4')

fig.suptitle('Reasons vs. Probability')

# plt.xlabel('Age')
# plt.ylabel('Probability')
```

```python
x = predicted_df['Transportation Expense']
children = predicted_df['Children']**2*100
y = predicted_df['Probability']*100

plt.scatter(x, y, s=children)

plt.title('Transportation Expense and Childrenvs. Probability')
plt.xlabel('Transportation Expense')
plt.ylabel('Probability')

plt.show()
```
