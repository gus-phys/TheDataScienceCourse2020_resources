## Import the releant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('datasets/iris-dataset.csv')
data.head()
```

```python
data.describe()
```

```python
fig, axes = plt.subplots(1, 4, figsize = (15,5))
sns.distplot(data['sepal_length'], ax = axes[0])
sns.distplot(data['sepal_width'], ax = axes[1])
sns.distplot(data['petal_length'], ax = axes[2])
sns.distplot(data['petal_width'], ax = axes[3])
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))
ax1.scatter(data['sepal_length'], data['sepal_width'])
ax1.set_xlabel('Sepal length')
ax1.set_ylabel('Sepal widith')
ax2.scatter(data['petal_length'], data['petal_width'])
ax2.set_xlabel('Petal length')
ax2.set_ylabel('Petal widith')
plt.show()
```
